```python
filepath2 = "Data Files/Cleaned Files/NIH Data Alcohol Consumption.csv"
filepath3 = "Data Files/Cleaned Files/NIH Data FIPS.csv"
filepath4 = "Data Files/Cleaned Files/NIH Beverage Codes.csv"

df_nih = pd.read_csv(filepath2)
df_beverage_codes = pd.read_csv(filepath4)
df_fips_codes = pd.read_csv(filepath3)

```

```python
df_nih.head()
```

```python
df_nih_merged = pd.merge(df_nih, df_beverage_codes, how="left", left_on = 'Type of beverage', right_on = 'Beverage ID')
df_nih_merged = pd.merge(df_nih_merged, df_fips_codes, how="left", left_on = 'FIPS', right_on = 'FIPS')
```

```python
df_nih_merged.head()
```

```python
# Create subsets for year

nih_1980 = df_nih_merged.loc[(df_nih_merged['Year'] == 1980) & (df_nih_merged['FIPS'] < 60)]
nih_1985 = df_nih_merged.loc[(df_nih_merged['Year'] == 1985) & (df_nih_merged['FIPS'] < 60)]
nih_1990 = df_nih_merged.loc[(df_nih_merged['Year'] == 1990) & (df_nih_merged['FIPS'] < 60)]
nih_1995 = df_nih_merged.loc[(df_nih_merged['Year'] == 1995) & (df_nih_merged['FIPS'] < 60)]
nih_2000 = df_nih_merged.loc[(df_nih_merged['Year'] == 2000) & (df_nih_merged['FIPS'] < 60)]
nih_2005 = df_nih_merged.loc[(df_nih_merged['Year'] == 2005) & (df_nih_merged['FIPS'] < 60)]
nih_2010 = df_nih_merged.loc[(df_nih_merged['Year'] == 2010) & (df_nih_merged['FIPS'] < 60)]
nih_2014 = df_nih_merged.loc[(df_nih_merged['Year'] == 2014) & (df_nih_merged['FIPS'] < 60)]
```

```python
# Get States Data from Substance Abuse DataFrame
df_states_mental_substance_use = df_Mental_substance_use.loc[df_Mental_substance_use['FIPS'] < 60]
```

```python
df_states_mental_substance_use.head()
```

```python
# Create new dataframe for substance abuse for each year

nih_1980_merged_substance = pd.merge(nih_1980, df_states_mental_substance_use, how = 'left', left_on = 'State', right_on = 'Location' )
nih_1985_merged_substance = pd.merge(nih_1985, df_states_mental_substance_use, how = 'left', left_on = 'State', right_on = 'Location' )
nih_1990_merged_substance = pd.merge(nih_1990, df_states_mental_substance_use, how = 'left', left_on = 'State', right_on = 'Location' )
nih_1995_merged_substance = pd.merge(nih_1995, df_states_mental_substance_use, how = 'left', left_on = 'State', right_on = 'Location' )
nih_2000_merged_substance = pd.merge(nih_2000, df_states_mental_substance_use, how = 'left', left_on = 'State', right_on = 'Location' )
nih_2005_merged_substance = pd.merge(nih_2005, df_states_mental_substance_use, how = 'left', left_on = 'State', right_on = 'Location' )
nih_2010_merged_substance = pd.merge(nih_2010, df_states_mental_substance_use, how = 'left', left_on = 'State', right_on = 'Location' )
nih_2014_merged_substance = pd.merge(nih_2014, df_states_mental_substance_use, how = 'left', left_on = 'State', right_on = 'Location' )
```

```python
# Grab only needed columns (don't forget to change year number in Mortality Rate column)
nih_1980_merged_substance = nih_1980_merged_substance[['Year','FIPS_x','Type of beverage', 'Gallons of ethanol per capita age 21 and older','Mortality Rate, 1980*','Location']]
nih_1985_merged_substance = nih_1985_merged_substance[['Year','FIPS_x','Type of beverage', 'Gallons of ethanol per capita age 21 and older','Mortality Rate, 1985*','Location']]
nih_1990_merged_substance = nih_1990_merged_substance[['Year','FIPS_x','Type of beverage', 'Gallons of ethanol per capita age 21 and older','Mortality Rate, 1990*','Location']]
nih_1995_merged_substance = nih_1995_merged_substance[['Year','FIPS_x','Type of beverage', 'Gallons of ethanol per capita age 21 and older','Mortality Rate, 1995*','Location']]
nih_2000_merged_substance = nih_2000_merged_substance[['Year','FIPS_x','Type of beverage', 'Gallons of ethanol per capita age 21 and older','Mortality Rate, 2000*','Location']]
nih_2005_merged_substance = nih_2005_merged_substance[['Year','FIPS_x','Type of beverage', 'Gallons of ethanol per capita age 21 and older','Mortality Rate, 2005*','Location']]
nih_2010_merged_substance = nih_2010_merged_substance[['Year','FIPS_x','Type of beverage', 'Gallons of ethanol per capita age 21 and older','Mortality Rate, 2010*','Location']]
nih_2014_merged_substance = nih_2014_merged_substance[['Year','FIPS_x','Type of beverage', 'Gallons of ethanol per capita age 21 and older','Mortality Rate, 2014*','Location']]
```

```python
# Rename Mortality Rate column
nih_1980_merged_substance.rename(columns={"Mortality Rate, 1980*": "Mortality Rate"}, inplace = True)
nih_1985_merged_substance.rename(columns={"Mortality Rate, 1985*": "Mortality Rate"}, inplace = True)
nih_1990_merged_substance.rename(columns={"Mortality Rate, 1990*": "Mortality Rate"}, inplace = True)
nih_1995_merged_substance.rename(columns={"Mortality Rate, 1995*": "Mortality Rate"}, inplace = True)
nih_2000_merged_substance.rename(columns={"Mortality Rate, 2000*": "Mortality Rate"}, inplace = True)
nih_2005_merged_substance.rename(columns={"Mortality Rate, 2005*": "Mortality Rate"}, inplace = True)
nih_2010_merged_substance.rename(columns={"Mortality Rate, 2010*": "Mortality Rate"}, inplace = True)
nih_2014_merged_substance.rename(columns={"Mortality Rate, 2014*": "Mortality Rate"}, inplace = True)
```

```python

# Concatenate all year dataframes
frames_substance = [nih_1980_merged_substance, 
                    nih_1985_merged_substance,
                    nih_1990_merged_substance,
                    nih_1995_merged_substance,
                    nih_2000_merged_substance,
                    nih_2005_merged_substance,
                    nih_2010_merged_substance,
                    nih_2014_merged_substance]

df_concat_substance = pd.concat(frames_substance)

df_concat_substance
```

```python
df_concat_substance = pd.merge(df_concat_substance,df_beverage_codes, how='left', left_on ='Type of beverage', right_on = 'Beverage ID')
```

```python
df_concat_substance_clean = df_concat_substance['Mortality Rate'].str.split(' ').str.get(0).astype(float)
```

```python
df_concat_substance['Mortality Rate'] = df_concat_substance_clean
```

```python
df_concat_substance
```

```python
# Isolate rows for all beverages
df_concat_substance4 = df_concat_substance.loc[df_concat_substance['Beverage ID'] == 4]
df_concat_substance1 = df_concat_substance.loc[df_concat_substance['Beverage ID'] == 1]
df_concat_substance2 = df_concat_substance.loc[df_concat_substance['Beverage ID'] == 2]
df_concat_substance3 = df_concat_substance.loc[df_concat_substance['Beverage ID'] == 3]
```

```python
df_concat_substance4.describe()
```

```python
# Remove nulls

df_concat_substance4.dropna(inplace=True)
df_concat_substance1.dropna(inplace=True)
df_concat_substance2.dropna(inplace=True)
df_concat_substance3.dropna(inplace=True)
```

```python
df_concat_substance2.describe()
```

```python
df_concat_substance2
```

```python
x_values4 = df_concat_substance4['Gallons of ethanol per capita age 21 and older']
y_values4 = df_concat_substance4['Mortality Rate']

x_values1 = df_concat_substance1['Gallons of ethanol per capita age 21 and older']
y_values1 = df_concat_substance1['Mortality Rate']

x_values2 = df_concat_substance2['Gallons of ethanol per capita age 21 and older']
y_values2 = df_concat_substance2['Mortality Rate']

x_values3 = df_concat_substance3['Gallons of ethanol per capita age 21 and older']
y_values3 = df_concat_substance3['Mortality Rate']
```

```python
plt.scatter(x_values4,y_values4)
plt.xlabel('Gallons of ethanol per capita age 21 and older')
plt.ylabel('Mortality Rate')
```

```python
corr4= st.pearsonr(x_values4,y_values4)
corr1= st.pearsonr(x_values1,y_values1)
corr2= st.pearsonr(x_values2,y_values2)
corr3= st.pearsonr(x_values3,y_values3)

(slope, intercept, rvalue, pvalue, stderr) = st.linregress(x_values2, y_values2)
regress_values2 = x_values2 * slope + intercept
line_eq2 = "y = " + str(round(slope,3)) + "x + " + str(round(intercept,2))




print(f"The correlation between ethanol consumption and mortality rate for all beverages is {corr4}")
print(f"The correlation between ethanol consumption and mortality rate for spirits is {corr1}")
print(f"The correlation between ethanol consumption and mortality rate for wine is {corr2}")
print(f"The correlation between ethanol consumption and mortality rate for beer is {corr3}")
```

```python
plt.scatter(x_values2,y_values2)
plt.plot(x_values2,regress_values2,"r-")
plt.annotate(line_eq2,(7000,25),fontsize=12,color="red")
plt.xlabel('Gallons of ethanol per capita age 21 and older')
plt.ylabel('Mortality Rate')
```
